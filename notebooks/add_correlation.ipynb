This notebook demonstrates how to make add correlation values to scatter plot using gpplot.

```python
# for development 
from gpplot_dev import add_gpplot_path
add_gpplot_path()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from gpplot import point_densityplot, calculate_correlation, add_correlation
import seaborn as sns
```

We pass the scatterplot and scatterplot data to add correlations

```python
nsamps = 20000
data = pd.DataFrame({'x': np.random.normal(size = nsamps)}, index = range(nsamps))
data['y'] = 2*data['x'] + np.random.normal(size = nsamps)
ax = point_densityplot(data, 'x', 'y')
add_correlation(ax, data, 'x', 'y')
```

We can change the location of the correlation

```python
ax = point_densityplot(data, 'x', 'y')
add_correlation(ax, data, 'x', 'y', loc = 'lower right')
```

Or the color and size of the correlation text

```python
ax = point_densityplot(data, 'x', 'y')
add_correlation(ax, data, 'x', 'y', size = 14, color = 'blue')
```

We can also use spearman correlation

```python
ax = point_densityplot(data, 'x', 'y')
add_correlation(ax, data, 'x', 'y', method = 'spearman')
```

# Pandas Barplot

Create a barplot using pandas plot functionality. While seaborn has a barplot function, it does not allow for stacked barplots, while pandas does. 

```python
import seaborn as sns
import pandas as pd
import gpplot
```

```python
mpg = sns.load_dataset('mpg')
mpg_summary = (mpg.groupby(['model_year', 'origin'])
               .agg({'mpg': 'mean'})
               .reset_index())
```

```python
ax = gpplot.pandas_barplot(mpg_summary, 'model_year', 'origin', 'mpg')
```

We can change the order of hues and x values

```python
hue_order = ['japan', 'usa', 'europe']
x_order = list(range(70, 83))[::-1]
ax = gpplot.pandas_barplot(mpg_summary, 'model_year', 'origin', 'mpg', hue_order=hue_order, x_order=x_order)
```

We can plot a vertical bar plot

```python
ax = gpplot.pandas_barplot(mpg_summary, 'model_year', 'origin', 'mpg', horizontal=False)
```

We can also change the aesthetics of the barplot

```python
ax = gpplot.pandas_barplot(mpg_summary, 'model_year', 'origin', 'mpg', 
                           color=sns.color_palette('Set2'), edgecolor = 'black')
```

Or make the bars very skinny

```python
ax = gpplot.pandas_barplot(mpg_summary, 'model_year', 'origin', 'mpg', 
                           color=sns.color_palette('Set2'), edgecolor = 'black', width = 0.2)
```

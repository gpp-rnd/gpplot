# Point Density Plot

This notebook demonstrates how to make a point density plot using gpplot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gpplot
```

```python
nsamps = 20000
data = pd.DataFrame({'x': np.random.normal(size = nsamps)}, index = range(nsamps))
data['y'] = 2*data['x'] + np.random.normal(size = nsamps)
ax = gpplot.point_densityplot(data, 'x', 'y')
```

We can change the colormap of the plot

```python
ax = gpplot.point_densityplot(data, 'x', 'y', palette='plasma')
```

We can add a legend for the color mapping

```python
gpplot.point_densityplot(data, 'x', 'y', legend = 'brief')
```

We can change the resolution for estimating the point density

```python
gpplot.point_densityplot(data, 'x', 'y', bins = [5,5])
```

```python
gpplot.point_densityplot(data, 'x', 'y', bins = [50,50])
```

We can make the plot square by using the "ax" argument

```python
fig, ax = plt.subplots(figsize=(4,4))
ax = gpplot.point_densityplot(data, 'x', 'y', ax = ax)
```

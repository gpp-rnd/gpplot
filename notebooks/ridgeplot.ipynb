This notebook demonstrates how to make overlapping kde plots ("ridgeplots") using gpplot. The code for ridgeplots was based on the example found here: https://seaborn.pydata.org/examples/kde_ridgeplot

```python
import gpplot
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
g = gpplot.ridgeplot(iris, 'sepal_width', 'species')
```

We can modify the position of the label text with the text_xpos, text_ypos, text_ha, text_va arguments.

```python
# Move the text to the right
g = gpplot.ridgeplot(iris, 'sepal_width', 'species', text_xpos=1, text_ha='right')
```

We can also change the opacity of the kde plots by changing alpha.

```python
# Really light
g = gpplot.ridgeplot(iris, 'sepal_width', 'species', alpha = 0.5)
```

```python
# Really dark
g = gpplot.ridgeplot(iris, 'sepal_width', 'species', alpha = 1)
```

To modify the overlap of the subplots, we make changes to the FacetGrid after the ridgeplots function returns a FacetGrid.

```python
# More overlap
g = gpplot.ridgeplot(iris, 'sepal_width', 'species')
g.fig.subplots_adjust(hspace=-0.5)
```

```python
# Less overlap
g = gpplot.ridgeplot(iris, 'sepal_width', 'species',)
g.fig.subplots_adjust(hspace=0)
```

Finally, we can also supply other parameters to FacetGrid via **kwargs. For example, we could specify the order of the kde plots via hue_order and row_order.

```python
# Just changing row order keeps the colors the same
g = gpplot.ridgeplot(iris, 'sepal_width', 'species', row_order=['versicolor','setosa','virginica'])
```

```python
# Change the colors by also specifying hue_order
g = gpplot.ridgeplot(iris, 'sepal_width', 'species', row_order=['versicolor','setosa','virginica'], hue_order=['versicolor','setosa','virginica'])
```

We can also change the color palette with palette

```python
# Paired color palette
g = gpplot.ridgeplot(iris, 'sepal_width', 'species', palette = sns.color_palette('Paired'))
```

```python
# Set 2 color palette
g = gpplot.ridgeplot(iris, 'sepal_width', 'species', palette = sns.color_palette('Set2'))
```

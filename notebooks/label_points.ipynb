# Label Points

Here we demonstrate labelling points in a scatter plot using gpplot

```python
import seaborn as sns
import gpplot
```

```python
mpg = sns.load_dataset('mpg')
ax = sns.scatterplot(data = mpg, x = 'weight', y = 'mpg')
label = ['hi 1200d', 'ford f250', 'chevy c20', 'oldsmobile omega']
gpplot.label_points(mpg, 'weight', 'mpg', label, 'name')
```

We can change the appearance of the arrows

```python
ax = sns.scatterplot(data = mpg, x = 'weight', y = 'mpg')

gpplot.label_points(mpg, 'weight', 'mpg', label, 'name', 
             arrowstyle='-', arrow_color='red',arrow_lw=2)
```

And the appearance of text

```python
ax = sns.scatterplot(data = mpg, x = 'weight', y = 'mpg')
gpplot.label_points(mpg, 'weight', 'mpg', label, 'name', 
             size = 12, style = 'italic')
```
